```python
def square_graph_code_generator(row_size, col_size):
    
    print("module graph(")
    print("    input clk,")
    print("    input rst,")
    print("    input ["+ str(row_size*col_size - 1) +":0] init,")
    for m in range(row_size*col_size):
        print("    output state"+ str(m) +",")
    print("    );")
    print()
    print()
    
    x = row_size + 2;
    y = col_size + 2;
    
    count = 0
    for z in range(y):
        for j in range(x):
            print("    reg out"+ str(count) +";")
            count += 1

    print()
    
    count = 0
    for i in range(y):
        for j in range(x):
            if (i==0):
                if (j==0):
                    print("    CELL cell_"+ str(count) +"(clk, rst, init["+ str(count) +"], 1'b0, out"+ str(count+1) +", out"+ str(count+x) +", 1'b0, out"+ str(count) +");")
                elif (j==(x-1)):
                    print("    CELL cell_"+ str(count) +"(clk, rst, init["+ str(count) +"], 1'b0, 1'b0, out"+ str(count+x) +", out"+ str(count-1) +", out"+ str(count) +");")
                else:
                    print("    CELL cell_"+ str(count) +"(clk, rst, init["+ str(count) +"], 1'b0, out"+ str(count+1) +", out"+ str(count+x) +", out"+ str(count-1) +", out"+ str(count) +");")
            elif (i==(y-1)):
                if (j==0):
                    print("    CELL cell_"+ str(count) +"(clk, rst, init["+ str(count) +"], out"+ str(count-x) +", out"+ str(count+1) +", 1'b0, 1'b0, out"+ str(count) +");")
                elif (j==(x-1)):
                    print("    CELL cell_"+ str(count) +"(clk, rst, init["+ str(count) +"], out"+ str(count-x) +", 1'b0, 1'b0, out"+ str(count-1) +", out"+ str(count) +");")
                else:
                    print("    CELL cell_"+ str(count) +"(clk, rst, init["+ str(count) +"], out"+ str(count-x) +", out"+ str(count+1) +", 1'b0, out"+ str(count-1) +", out"+ str(count) +");")
            else:
                if (j==0):
                    print("    CELL cell_"+ str(count) +"(clk, rst, init["+ str(count) +"], out"+ str(count-x) +", out"+ str(count+1) +", out"+ str(count+x) +", 1'b0, out"+ str(count) +");")
                elif(j==(x-1)):
                    print("    CELL cell_"+ str(count) +"(clk, rst, init["+ str(count) +"], out"+ str(count-x) +", 1'b0, out"+ str(count+x) +", out"+ str(count-1) +", out"+ str(count) +");")
                else:
                    print("    CELL cell_"+ str(count) +"(clk, rst, init["+ str(count) +"], out"+ str(count-x) +", out"+ str(count+1) +", out"+ str(count+x) +", out"+ str(count-1) +", out"+ str(count) +");")
            count += 1
            
    print()
    
    count = 0
    count_used = 0
    for i in range(y):
        for j in range(x):
            if ((i>0) and (i<(y-1))):
                if ((j>0) and (j<(x-1))):
                    print("    assign state"+ str(count_used) +" = out"+ str(count) +";")
                    count_used += 1
            count += 1
                    
    print()
    print("endmodule")
            
```

```python
square_graph_code_generator(10,10)
```

```python
def graph_tb_code_generator(row_size, col_size):
    print("module graph_tb();")
    for i in range(row_size*col_size):
        print("    wire state"+ str(i) +";")
    print("    reg clk, rst;")
    print("    reg ["+ str(row_size*col_size - 1) +":0] init;")
    print()
    print()
    print("    graph UUT(")
    for i in range(row_size*col_size):
        print("        .state"+ str(i) +"(state"+ str(i) +"),")
    print("        .clk(clk),")
    print("        .rst(rst),")
    print("        .init(init)")
    print("    );")
    print()
    print()
    print("    initial begin")
    print("        clk = 0;")
    print("        rst = 0;")
    print("        forever")
    print("            #10 clk = ~clk;")
    print("    end")
    print()
    print("    initial begin")
    print("        #5{rst, init} = {1'b1, "+ str(row_size*col_size) +"'h};") #Have to add custom starting condition after generation
    print("        #10 rst = 1'b0;")
    
    print()
    count = 0
    for i in range(col_size):
        for j in range(row_size):
            if (j==0):
                print('        $monitor("%b, ', end=' ')
            elif (j==(row_size-1)):
                print('%b", ', end=' ')
            else:
                print("%b, ", end=' ')
        for k in range(row_size):
            if (k==(row_size-1)):
                print("state"+ str(count) +");")
            else:
                print("state"+ str(count) +", ", end='')
            count += 1
            
    print("    end")
    print("endmodule")
```

```python
graph_tb_code_generator(10,10)
```

```python
def create_hex_representation(num_elements,live_cells):
    bin_list = []
    for i in range(num_elements):
        for cell in live_cells:
            if (cell==i):
                bin_list.append(1)
            else:
                bin_list.append(0)
                
    #padding bin_list
    for i in range(4-(len(bin_list)%4)):
        bin_list.append(0)
    
    #converting to hex
    count = 0
    hex_list = []
    for j in range(len(bin_list)/4):
        sing_hex = []
        for k in range(4):
            sing_hex.append(bin_list[count])
            count += 1
        if(sing_hex==[0,0,0,0]):
            hex_list.append('0')
        elif(sing_hex==[1,0,0,0]):
            hex_list.append('1')
        elif(sing_hex==[0,1,0,0]):
            hex_list.append('2')
        elif(sing_hex==[1,1,0,0]):
            hex_list.append('3')
        elif(sing_hex==[0,0,1,0]):
            hex_list.append('4')
        elif(sing_hex==[1,0,1,0]):
            hex_list.append('5')
        elif(sing_hex==[0,1,1,0]):
            hex_list.append('6')
        elif(sing_hex==[1,1,1,0]):
            hex_list.append('7')
        elif(sing_hex==[0,0,0,1]):
            hex_list.append('8')
        elif(sing_hex==[1,0,0,1]):
            hex_list.append('9')
        elif(sing_hex==[0,1,0,1]):
            hex_list.append('A')
        elif(sing_hex==[1,1,0,1]):
            hex_list.append('B')
        elif(sing_hex==[0,0,1,1]):
            hex_list.append('C')
        elif(sing_hex==[1,0,1,1]):
            hex_list.append('D')
        elif(sing_hex==[0,1,1,1]):
            hex_list.append('E')
        elif(sing_hex==[1,1,1,1]):
            hex_list.append('F')

    for num in reversed(hex_list):
        print(num, end=' ')
```

```python
create_hex_representation(100, [2,12,22,21,10])
```
